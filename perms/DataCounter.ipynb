```python
y = x.map(lambda k : [str(k[1]),str(k[2])])
```

```python
def getCounterValue(vals):
    c = int(vals[0],16)*pow(2,8) + int(vals[1],16)
    return c
    
```

```python
k = ["0A", "7D"];
getCounterValue(k)
```

```python
s = y.map(lambda k: getCounterValue(k))
```

```python
s
```

```python
u = s.to_frame().diff()
```

```python
m = u.mean()
print(m)
```
